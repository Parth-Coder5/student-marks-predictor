## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Start Cleaning

```python
df = pd.read_csv('students.csv')
```

### Distribution of Final Marks (G3)

```python
plt.figure(figsize=(6,4))
sns.histplot(df['G3'],bins=20,kde=True)
plt.title('Distribution of Final Marks (G3)')
plt.xlabel('Final Marks')
plt.ylabel('Count')
```

### Study Time VS Marks (G3)

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='studytime',y='G3',data=df)
plt.title('Study VS Marks')
plt.xlabel('Study Time')
plt.ylabel('Final Marks (G3)')
```

### Absence in School VS Marks (G3)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='absences',y='G3',data=df)
plt.title('Absent in School VS Marks (G3)')
plt.xlabel('Absents')
plt.ylabel('Final Marks (G3)')
```
